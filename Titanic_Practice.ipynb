```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('Titanic-Dataset.csv')
df.head()
```

# Categorical Data

## a) Countplot 

```python
sns.countplot(x='Survived', data=df, hue='Survived')
```

```python
sns.countplot(x='Pclass', data=df, hue='Pclass')
```

## b) Pie Chart

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%.2f')
```

# Numerical Data

## a) Histogram

```python
plt.hist(df['Age'])
```

## b) Displot

```python
sns.displot(df['Age'])
```

## c) Boxplot

```python
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['Age'])
```
