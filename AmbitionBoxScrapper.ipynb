```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

final=pd.DataFrame()
dfs=[]
TEST_RUN=5 #for full run, TEST_RUN=500
for j in range (1,TEST_RUN+1):
    url='https://www.ambitionbox.com/list-of-companies?campaign=homepage_explore_companies_widget&page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage=requests.get(url,headers=headers).text
    
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all(class_='companyCardWrapper__primaryInformation')
    name=[]
    rating=[]
    rating_count=[]
    about=[]
    values=[]
    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find(class_='rating_text rating_text--md').text.strip())
        rating_count.append(i.find(class_='companyCardWrapper__companyRatingCount').text.strip().strip("()"))
        about.append(i.find(class_='companyCardWrapper__interLinking').text.strip())
        values.append(i.find(class_='companyCardWrapper__ratingValues').text.strip()
                     if i.find(class_='companyCardWrapper__ratingValues') else " ")
    d={'Name': name, 'Rating': rating, 'Rating Count': rating_count, 'About': about, 'Values':values}
    df=pd.DataFrame(d)
    dfs.append(df)
final = pd.concat(dfs, ignore_index=True)
final
```
