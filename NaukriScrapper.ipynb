```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
driver=webdriver.Safari()
wait=WebDriverWait(driver,10)
final=pd.DataFrame()

job_title=[]
company_name=[]
rating=[]
reviews=[]
exp_wdth=[]
job_desc=[]
```

```python

TEST_RUN=4 #For full run, TEST_RUN=457
for page in range (1,TEST_RUN+1):
    url = "https://www.naukri.com/jobs-in-jaipur-{}".format(page)
    driver.get(url)
    wait.until(EC.presence_of_element_located((By.CLASS_NAME, "cust-job-tuple")))
    html=driver.page_source
    soup=BeautifulSoup(html, 'lxml')
    company=soup.find_all(class_='cust-job-tuple')

    for i in company:
        job_title.append(i.find(class_='title').text.strip() if i.find(class_='title') else np.nan)
        company_name.append(i.find(class_='comp-name').text.strip() if i.find(class_='comp-name') else np.nan)
        rating.append(i.find(class_='main-2').text.strip() if i.find(class_='main-2') else np.nan)
        reviews.append(i.find(class_='review ver-line').text.strip().split()[0] if i.find(class_='review ver-line') else np.nan)
        exp_wdth.append(i.find(class_='expwdth').text.strip() if i.find(class_='expwdth') else np.nan)
        job_desc.append(i.find(class_='job-desc').text.strip() if i.find(class_='job-desc') else np.nan)
        
final=pd.DataFrame({"Job Title": job_title, "Company Name": company_name, "Rating": rating, "Reviews": reviews, "Experience": exp_wdth, "Job Description": job_desc})
final=final.drop_duplicates().reset_index(drop=True)
final
```
